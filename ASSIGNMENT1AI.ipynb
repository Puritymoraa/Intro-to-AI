```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States')
page.status_code
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', {'class': 'wikitable'})

# Extract column headers
headers = [header.text.strip().replace('\n', '') for header in table.find_all('th')]
```

```python
#table = soup.find_all('table')[0]
print(table)
```

```python
world_titles = table.find_all('th')
world_titles
```

```python
world_table_titles = [title.text.strip().replace('\n','').replace('[a]', '').replace('[b]','').replace('[16]', '').replace('[15]','').replace('[14]','') for title in world_titles]
print(world_table_titles)
```

```python
limited_titles = world_table_titles[:8]
limited_titles
```

```python
rows = table.find_all('tr')
data = []
for row in rows[1:]:
    row_data = row.find_all(['th', 'td'])
    individual_row_data = [data.text.strip().replace('\n\n', ' - ').replace('\n', ' ')
                          for data in row_data][:8]
    data.append(individual_row_data)


president_data = pd.DataFrame(data=data, columns=limited_titles)
president_data.dropna(inplace=True)
president_data
```

```python
#QUESTION2. How many presidents are on the web page?
president_names = president_data['Name(Birth–Death)'].tolist()
unique_president_names = set(president_names)
print("Number of presidents on the web page: ", len(unique_president_names))
```

```python
#QUESTION3. How many presidencies ever existed in the United states
total_presidencies = president_data['Name(Birth–Death)'].count()
print("Number of presidents on the web page: ", total_presidencies)
```

```python
#QUESTION4.
# Convert the 'Term' column to integer values
president_data['Term'] = president_data['Term'].astype(int)

# Filter for presidents who served more than one term
multiple_term_presidents = president_data[president_data['Term'] > 1]
```

```python
#QUESTION5. Which president was in office longer than others and when did he get in office?
#This code fulfills the following: The code picks the name of the president, 
#the number of terms he was in office and also extracts the times he went in office
president_data['Start Term'] = pd.to_datetime(president_data['Term'].str.split('–').str[0], errors='coerce')
president_data['End Term'] = pd.to_datetime(president_data['Term'].str.split('–').str[1].replace('Incumbent', pd.Timestamp.now().strftime('%Y-%m-%d')), errors='coerce')
```

```python
president_data['Term Duration'] = president_data['End Term'] - president_data['Start Term']
longest_serving_president = president_data.loc[president_data['Term Duration'].idxmax()]
```

```python
print("President who was in office longer than others and when he got into office:")
print(f"Name: {longest_serving_president['Name(Birth–Death)']}")
print(f"Start Term: {longest_serving_president['Start Term'].date()}")
print(f"Duration in Office: {longest_serving_president['Term Duration'].days} days")
```

```python
#QUESTION6. Which presidents were elected after serving as vice presidents?
#The code should fulfills the following: The code manages to pick the names of the 
#presidents who once served as vice presidents and the periods they served as vice president and president

# Normalize the term information
president_data['Term in office'] = president_data['Term'].str.replace('–', '-')

# Identify presidents who were elected after serving as vice presidents
presidents_as_vp = president_data.groupby('Vice President').filter(lambda x: x['Name(Birth–Death)'].nunique() > 1)

# Extract the required information: name, periods served as vice president and president
presidents_info = presidents_as_vp[['Name(Birth–Death)', 'Term in office', 'Vice President']]

# Print the information
print(presidents_info)
```

```python
#QUESTION7. How many presidents were never elected, and what are their names
#The code managed to pick the names of the presidents who once served as senators 
#and the periods they served as senator and president

# Define a list of values indicating indirect election
indirect_election_values = ["–", "Died in office", "Resigned from office"]
```

```python
# Identify presidents who were never elected
never_elected = president_data[
    (president_data['Election'].apply(lambda x: any(value in x for value in indirect_election_values)))
]['Name(Birth–Death)'].unique()
```

```python
# Print the results
print(f"Number of presidents never elected: {len(never_elected)}")
print("Names of presidents never elected:")
for name in never_elected:
    print(f"\t- {name}")
```

```python
#QUESTION8. How many presidents have the same forenames and/or surnames and who are they?
names = president_data['Name(Birth–Death)'].str.split(' ')
first_names = names.apply(lambda x: x[0])
last_names = names.apply(lambda x: x[-1])
```

```python
common_first_names = first_names[first_names.duplicated(keep=False)].unique()
common_last_names = last_names[last_names.duplicated(keep=False)].unique()
print("Presidents with common forenames:")
print(president_data[president_data['Name'].str.split(' ').apply(lambda x: x[0]).isin(common_first_names)]['Name'].unique())
print("Presidents with common surnames:")
print(president_data[president_data['Name'].str.split(' ').apply(lambda x: x[-1]).isin(common_last_names)]['Name'].unique())
```
